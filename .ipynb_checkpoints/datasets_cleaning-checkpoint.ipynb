# Cleaning and unifying data sets

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_red = pd.read_csv('data/winequality-red.csv',sep='\;', engine='python')
df_red.head()
```

```python
df_red.columns = ['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar', 'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']
```

```python
df_red.head()
```

```python
df_red = df_red.assign(type='red')
```

```python
df_red.head()
```

```python
df_red.to_csv('redwine-qa.csv', index=False)
```

```python
df_white = pd.read_csv('data/winequality-white.csv',sep='\;', engine='python')
df_white.head()
```

```python
df_white.columns = ['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar', 'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']
```

```python
df_white.head()
```

```python
df_white = df_white.assign(type='white')
```

```python
df_white.head()
```

```python
df_white.to_csv('whitewine-qa.csv', index=False)
```

```python
dataframes = [df_red, df_white]
df = pd.concat(dataframes, sort='False')
```

```python
df.head()
```

```python
df.describe()
```

```python
import seaborn as sns

sns.heatmap(df.isnull())
```

```python
df.info()
```

```python
df.to_csv('wine-qa.csv', index=False)
```

# Data distribution

```python
df = pd.read_csv('data/wine-qa.csv')
```

```python
fig, ax = plt.subplots(nrows=3, ncols=4, figsize=(12,12))
fig.tight_layout(pad=2, h_pad=4)

ax[0,0].hist(df['alcohol'], color='#D1AFE9', rwidth=0.9)
ax[0,0].set_title('Alcohol')
ax[0,0].set_xlabel('vol.%')
ax[0,0].set_ylabel('Count')
    
ax[0,1].hist(df['chlorides'], color='#D1AFE9', rwidth=0.9)
ax[0,1].set_title('Chlorides')
ax[0,1].set_xlabel('g(sodium chloride)/dm3')
ax[0,1].set_ylabel('Count')

ax[0,2].hist(df['citric_acid'], color='#D1AFE9', rwidth=0.9)
ax[0,2].set_title('Citric acid')
ax[0,2].set_xlabel('g/dm3')
ax[0,2].set_ylabel('Count')

ax[0,3].hist(df['density'], color='#D1AFE9', rwidth=0.9)
ax[0,3].set_title('Density')
ax[0,3].set_xlabel('g/cm3')
ax[0,3].set_ylabel('Count')

ax[1,0].hist(df['fixed_acidity'], color='#D1AFE9', rwidth=0.9)
ax[1,0].set_title('Fixed acidity')
ax[1,0].set_xlabel('g(tartaric acid)/dm3')
ax[1,0].set_ylabel('Count')

ax[1,1].hist(df['free_sulfur_dioxide'], color='#D1AFE9', rwidth=0.9)
ax[1,1].set_title('Free sulfur dioxide')
ax[1,1].set_xlabel('mg/dm3')
ax[1,1].set_ylabel('Count')

ax[1,2].hist(df['pH'], color='#D1AFE9', rwidth=0.9)
ax[1,2].set_title('pH')
ax[1,2].set_xlabel('units')
ax[1,2].set_ylabel('Count')

ax[1,3].hist(df['quality'], color='#D1AFE9', rwidth=0.9)
ax[1,3].set_title('Quality')
ax[1,3].set_xlabel('units')
ax[1,3].set_ylabel('Count')

ax[2,0].hist(df['residual_sugar'], color='#D1AFE9', rwidth=0.9)
ax[2,0].set_title('Residual sugar')
ax[2,0].set_xlabel('g/dm3')
ax[2,0].set_ylabel('Count')

ax[2,1].hist(df['sulphates'], color='#D1AFE9', rwidth=0.9)
ax[2,1].set_title('Sulphates')
ax[2,1].set_xlabel('g(potassium sulphate)/dm3')
ax[2,1].set_ylabel('Count')

ax[2,2].hist(df['total_sulfur_dioxide'], color='#D1AFE9', rwidth=0.9)
ax[2,2].set_title('Total sulfur dioxide')
ax[2,2].set_xlabel('mg/dm3')
ax[2,2].set_ylabel('Count')

ax[2,3].hist(df['volatile_acidity'], color='#D1AFE9', rwidth=0.9)
ax[2,3].set_title('Volatile acidity')
ax[2,3].set_xlabel('g(acetic acid)/dm3')
ax[2,3].set_ylabel('Count')

plt.show()
```

```python
fig.savefig("images/data_dist.png", bbox_inches='tight')
```
