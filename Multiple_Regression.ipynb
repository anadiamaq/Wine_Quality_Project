# Multiple Regression

Linear regression models the relationship between a dependent variable and two or more independent variables. In this case, there are a total of 12 variables. 11 variables are assumed as predictors o independent variables and one (quality) as dependet varibale.
This model helps to know which variables are most related to quality (dependent variable).

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from scipy import stats
```

```python
wine_qa = pd.read_csv('data/wine-qa.csv')
wine_qa.head()
```

```python
y = wine_qa['quality']
X = wine_qa.drop(columns=['quality','type'])
```

```python
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
df_red = pd.read_csv('data/redwine-qa.csv')
df_red.head()
```

```python
y = df_red['quality']
X_red = df_red.drop(columns=['quality','type'])
```

```python
X3 = sm.add_constant(X_red)
est = sm.OLS(y, X3)
est2 = est.fit()
print(est2.summary())
```

The multiple regression for red wine variables shows that volatile acidity, chlorides, free sulfur dioxide, total sulfur dioxide, pH, sulphates and alcohol are the independent variables related to wine quality in red wines.

```python
df_white = pd.read_csv('data/whitewine-qa.csv')
df_white.head()
```

```python
y = df_white['quality']
X_white = df_white.drop(columns=['quality','type'])
```

```python
X4 = sm.add_constant(X_white)
est = sm.OLS(y, X4)
est2 = est.fit()
print(est2.summary())
```

The multiple regression for white wine variables shows that fixed acidity, volatile acidity, residual sugar, free sulfur dioxide, pH, sulphates and alcohol are the independent variables realated to wine quality in white wines.
