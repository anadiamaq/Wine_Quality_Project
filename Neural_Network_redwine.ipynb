```python
import pandas as pd
```

```python
df_red = pd.read_csv('data/redwine-qa.csv')
df_red.head()
```

```python
y = df_red['quality']
X = df_red.drop(columns=['quality','type','fixed_acidity','citric_acid', 'residual_sugar', 'density'])
```

```python
from sklearn.preprocessing import StandardScaler
prep = StandardScaler()
prep.fit(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from keras import regularizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(12, activation='relu', input_shape=(7,),kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(12, activation='relu'))
model.add(Dense(1, activation='relu'))
```

```python
model.compile(loss= 'mse',
              optimizer= 'adam',
              metrics=['mae'])
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=200, batch_size=128, verbose=2)
```

```python
model.fit(X, y)
```

```python
score = model.evaluate(X_test,y_test,batch_size=42,verbose=2)
score
```

```python
from matplotlib import pyplot
pyplot.title('Loss / Mean Squared Error')
pyplot.plot(history.history['loss'], label='train_loss')
pyplot.plot(history.history['val_loss'], label='train_val')
pyplot.legend()
pyplot.show()
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred.round()
```
